```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from os import listdir
```

```python
day3 = (listdir("data/ABC/spectrograms/2019003"))
sig_list = [x for x in day3 if 'sig' in x]
sig_list = sorted(sig_list)
sig_list = sig_list[1:]
day3 = sorted(day3)
day3 = day3[6:]
```

```python
for i in range(0,50):
    #show normal version
    folder = "data/ABC/spectrograms/2019003/"
    file = day3[i]
    filename = folder + file
    print(filename)
    im = cv2.imread(filename)
    RGB_img = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20, 10))
    ax1.axis('off')
    RGB_img = RGB_img[140:230, 60:-50, :] 
    ax1.imshow(RGB_img)
    
    #fixes bullshit opencv
    res = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    #crops image
    res = res[140:230, 60:-50, :]    
    #stops matplotlib from making bullshit color schemes
    plt.rc('image', cmap='gray')
    #apply gaussian blur
    #hoeizontal
    res = cv2.blur(res,(6,3))
    #Creates canny edges
    display = cv2.Canny(res, 90, 300)
    print(display.shape)
    ax2.imshow(display)
    break
```
